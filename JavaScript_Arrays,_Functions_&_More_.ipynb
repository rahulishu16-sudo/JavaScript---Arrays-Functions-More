5. Given an array numbers = [1, 2, 3, 4, 5], perform the following:

   1. Use a for loop to print each number.
   2. Use a while loop to print each number.
   3. Use a for-in loop to print each index.
   4. Use a for-of loop to print each number.

- This is how you can use different loop types in JavaScript to print numbers and indices from the array numbers = [1, 2, 3, 4, 5]:

     let numbers = [1, 2, 3, 4, 5];

     1. For loop to print each number:-

       for (let i = 0; i < numbers.length; i++) {

          console.log(numbers[i]);
       }

     2. While loop to print each number:-

       let j = 0;

       while (j < numbers.length) {

          console.log(numbers[j]);

          j++;

       }

     3. For-in loop to print each index:-

       for (let index in numbers) {

         console.log(index);

       }

     4. For-of loop to print each number:-

       for (let num of numbers) {

         console.log(num);

       }
